```python
from analysis import DataPreprocessor, ProphetForecaster
import pandas as pd

preprocessor = DataPreprocessor('test_data.csv')
prophet_data = preprocessor.prepare_for_prophet(
    preprocessor.df, 
    target_column='PRECTOTCORR', 
    year_column='YEAR', 
    doy_column='DOY'
)

forecaster = ProphetForecaster(
    seasonality_mode='additive',
    changepoint_prior_scale=0.05
)
forecaster.create_model(
    yearly_seasonality=True,
    weekly_seasonality=False
)

forecaster.fit(prophet_data)
forecast = forecaster.predict(periods=180, frequency='D', include_history=True)
print(forecast)
```
